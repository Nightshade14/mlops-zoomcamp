```python
def preprocess_df_to_xy(df, target="duration", dv = None):
	df["duration"] = df["tpep_dropoff_datetime"] - df["tpep_pickup_datetime"]
	df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)
	df = df[(df.duration >= 1) & (df.duration <= 60)]
	categorical_cols = ['PULocationID', 'DOLocationID']
	df[categorical_cols] = df[categorical_cols].astype(str)
	data_dicts = df[categorical_cols].to_dict(orient='records')
	if not dv:
		dv = feature_extraction.DictVectorizer()
		X = dv.fit_transform(data_dicts)
	else:
		X = dv.transform(data_dicts)
	y = df[target].values

	return (X,y,dv)
```

```python
df = pd.read_parquet("./data/yellow_tripdata_2023-01.parquet")
X_train, y_train, dv = preprocess_df_to_xy(df)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import root_mean_squared_error

y_pred = lr.predict(X_train)
ans5 = root_mean_squared_error(y_train, y_pred)
print(f"Answer5: {round(ans5,3)}")
```

```python
val_df = pd.read_parquet("./data/yellow_tripdata_2023-02.parquet")
X_val, y_val, _ = preprocess_df_to_xy(df=val_df, dv=dv)
```

```python
y_pred = lr.predict(X_val)
ans6 = root_mean_squared_error(y_val, y_pred)
print(f"Answer6: {round(ans6,3)}")
```
